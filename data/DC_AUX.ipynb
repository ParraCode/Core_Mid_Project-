```python
import pandas as pd 
```

```python
url_UCI = "https://opendata.ecdc.europa.eu/covid19/hospitalicuadmissionrates/csv/data.csv"
df_uci = pd.read_csv(url_UCI)
```

```python
df_uci.head(100)
```

```python
print((df_uci.isnull().sum()/len(df_uci))*100)
```

```python
df_uci['indicator'].unique()
```

```python
# Divido el dataset en 4 dataset por indicator y luego los uno en uno solo por columnas
df_uci['id-merge-country-date'] = df_uci['country']+df_uci['date']
df_uci = df_uci.drop(['year_week', 'source', 'url'], axis=1)
```

```python
# Hospital_Occupancy

filter_uci = df_uci['indicator'] == 'Daily hospital occupancy'

# DEFINO EL DATAFRAME
df_uci_1 = df_uci[filter_uci]
df_uci_1 = df_uci_1.rename(columns={'value':'Hospital_Occupancy'})
df_uci_1 = df_uci_1.drop(['indicator','date','country'], axis=1)

df_uci_1
```

```python
# ICU occupancy

filter_uci = df_uci['indicator'] == 'Daily ICU occupancy'

#DEFINO EL DATAFRAME
df_uci_2 = df_uci[filter_uci]
df_uci_2 = df_uci_2.rename(columns={'value':'ICU_Occupancy'})
df_uci_2 = df_uci_2.drop(['indicator','date','country'], axis=1)

df_uci_2
```

```python
df_uci = df_uci.drop(['indicator','value','country'], axis=1)
df_uci
```

```python
df_uci = pd.merge(df_uci, df_uci_1 , how='left', on='id-merge-country-date')
df_uci = pd.merge(df_uci, df_uci_2 , how='left', on='id-merge-country-date')
df_uci
```
